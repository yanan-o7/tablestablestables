### Программа


```python
import csv
import pickle

def load_table_csv(use):
    """
    Выводит данные csv файла

    Args:
      use (str): Имя файла

    Returns:
      list: Список с содержимым файла
    """
    try:
        with open(use, 'r', newline='') as file:
            data = list(csv.reader(file))
            return (data)
    except FileNotFoundError:
        raise FileNotFoundError('File not found')
    except Exception:
        raise Exception('Loading error')
print('Load csv:', load_table_csv('file_1.csv'))

def save_table_csv(use):
    """
    Изменяет данные csv файла

    Args:
      use (str): Имя файла

    Returns:
      list: Список с содержимым файла
    """
    try:
        with open(use, 'w', newline='', encoding='utf-8') as file:
            csv.writer(file).writerow(['y','o','u'])
            csv.writer(file).writerow(['cha','ng','ed'])
            csv.writer(file).writerow(['some','th','ing'])
            csv.writer(file).writerow(['he','re','!!!'])
        with open(use, 'r', newline='', encoding='utf-8') as file:
            data = list(csv.reader(file))
            return (data)
    except Exception as e:
        raise Exception('Saving error')
print('Save csv:', save_table_csv('file_1.1.csv'))

def load_table_pickle(use):
    """
    Выводит данные pickle файла

    Args:
      use (str): Имя файла

    Returns:
      list: Список с содержимым файла
    """
    try:
        with open(use, 'rb') as file:
            return pickle.load(file)
    except FileNotFoundError:
        raise FileNotFoundError('File not found')
    except Exception:
        raise Exception('Loading error')
print('Load pickle:', load_table_pickle('file_2.pickle'))

new_data = [('y', 'o', 'u'),
('cha', 'ng', 'ed'),
('the', 'da', 'ta')]
def save_table_pickle(use):
    """
    Изменяет данные pickle файла

    Args:
      use (str): Имя файла

    Returns:
      list: Список с содержимым файла
    """
    try:
        with open(use, 'wb') as file:
            pickle.dump(new_data, file)
        with open(use, 'rb') as file:
            return pickle.load(file)        
    except Exception:
        raise Exception('Saving error')
print('Save pickle:', save_table_pickle('file_2.2.pickle'))

final = []
def save_table_txt(use):
    """
    Выводит данные txt файла

    Args:
      use (str): Имя файла

    Returns:
      list: Список с содержимым файла
    """
    try:
        with open(use) as file:
            for info in file:
                x,y,z = map(str,info.split())
                mini = [x,y,z]
                final.append(mini)
            return final
    except FileNotFoundError:
        raise FileNotFoundError('File not found')
    except Exception:
        raise Exception('Saving error')
print('Save txt:', save_table_txt('file_3.txt'))

def get_rows_by_number (use, start, stop, copy_table=0):
    """
    Выводит строки файла

    Args:
      use, start, stop, copy_table (str, int, int, bool): Имя файла, начало, конец отрезка, проверка на копирование

    Returns:
      list: Список с выбранными строками файла
    """
    try:
      return load_table_csv(use)[start:stop]
    except IndexError:
        raise IndexError('Wrong interval values')
    except Exception:
        raise Exception('An error has occurred')
print('Rows num:', get_rows_by_number('file_1.csv', 1, 2))

new_list=[]
def get_rows_by_index(use, *values, copy_table=0):
    """
    Выводит строки файла

    Args:
      use, values, copy_table (str, str, bool): Имя файла, значения, проверка на копирование

    Returns:
      list: Список с выбранными строками файла
    """
    try:
        new_list.append(load_table_csv(use)[0])
        new_list.append(load_table_csv(use)[1])
        new_list.append(load_table_csv(use)[2])
        new_list.append(load_table_csv(use)[3])
        rows = [row for row in new_list if row[0] in values]
        return rows
    except Exception:
       raise Exception('An error has occurred')
print('Rows id:', get_rows_by_index('file_1.csv', 'yanan'))

new_list=[]
def get_column_types(use):
    """
    Выводит типы столбцов файла

    Args:
      use (str): Имя файла

    Returns:
      dict: Словарь со столбцами и типом значений в них
    """
    try:
        new_list.append(type(load_table_csv(use)[0][0]))
        new_list.append(type(load_table_csv(use)[0][1]))
        new_list.append(type(load_table_csv(use)[0][2]))
        column=[0,1,2]
        types=dict(zip(column,new_list))
        return types
    except Exception:
        raise Exception('An error has occurred')
print('Get column type:', get_column_types('file_1.csv'))

new_list=[]
types={}
def set_column_types(use):
    """
    Выводит строки файла

    Args:
      use (str): Имя файла

    Returns:
      dict: Словарь со столбцами и типом значений в них
    """
    try:
        new_list.append(load_table_csv(use)[0][0])
        new_list.append(load_table_csv(use)[0][1])
        new_list.append(load_table_csv(use)[0][2])
        for value in new_list:
            if value.isdigit():
               value = 'int'
            elif value.lower() in ["true", "false"]:
                value = 'bool'
            else:
                value = 'str'
            for column in range(0,3):
                types.update(column=value)
        return types
    except Exception:
        raise Exception('An error has occurred')
print('Set column type:', set_column_types('file_1.csv'))

new_list = []
def get_values(use, column=0):
    """
    Выводит значения из файла

    Args:
      use, column (str, int): Имя файла, Индекс столбца

    Returns:
      list: Список с выбранным столбцом файла
    """
    try:
        new_list.append(load_table_csv(use)[0][column])
        new_list.append(load_table_csv(use)[1][column])
        new_list.append(load_table_csv(use)[2][column])
        new_list.append(load_table_csv(use)[3][column])
        return new_list
    except ValueError:
        raise ValueError('Column not found')
    except Exception:
        raise Exception('An error has occurred')
print('Get values:', get_values('file_1.csv', 0))

def get_value(use, column=0):
    """
    Выводит значение из файла

    Args:
      use, column (str, int): Имя файла, Индекс столбца

    Returns:
      str: значение из выбранного столбца файла
    """
    try:
        return load_table_csv(use)[0][column]
    except ValueError:
        raise ValueError('Column not found')
    except Exception:
        raise Exception('An error has occurred')
print('Get value:', get_value('file_1.csv', 0))

new_list=[]
def set_values(use, value, new_value, column=0):
    """
    Меняет значения из файла

    Args:
      use, value, new_value, column (str, int): Имя файла, начальное значение, изменённое значение, Индекс столбца

    Returns:
      list: Список с выбранным столбцом файла
    """
    try:
        new_list.append(load_table_csv(use)[0][column])
        new_list.append(load_table_csv(use)[1][column])
        new_list.append(load_table_csv(use)[2][column])
        new_list[new_list.index(value)]=new_value
        return new_list
    except Exception:
        raise Exception('An error has occurred')
print('Set values:', set_values('file_1.csv', 'yanan', 'not_yanan', 0))

new_list=[]
def set_value(use, value, new_value, column=0):
    """
    Меняет значение из файла

    Args:
      use, value, new_value, column (str, int): Имя файла, начальное значение, изменённое значение, Индекс столбца

    Returns:
      list: Список с выбранным столбцом файла
    """
    try:
        new_list.append(load_table_csv(use)[0][column])
        new_list[new_list.index(value)]=new_value
        return new_list
    except Exception:
        raise Exception('An error has occurred')
print('Set value:', set_value('file_1.csv', 'yanan', 'not_yanan', 0))

def print_table(use):
    """
    Выводит данные файла

    Args:
      use: Имя файла

    Returns:
      list: Список с содержимым файла
    """
    try:
        with open(use, 'r', encoding='utf-8') as file:
            lines = file.readlines()
        return lines
    except Exception:
        raise Exception('An error has occurred')
print('Print txt:', print_table('file_3.txt'))
```
